```python
# challenge 1
# Proram ma za zadanie sprawdzać z jakim województwem mamy do czynienia gdy wyświelta się numer tel.
# Sprawdza również czy numer jest prawidołowy 
# program posiada możliwosć wypisania słownego numeru
class NumerTelefonu:

    def __init__(self, numer_tel):
        self.numer = numer_tel
        self.slownie = {1: "jeden", 2: "dwa", 3: "trzy", 4: "cztery", 5: "pięć", 6: "sześć", 7: "siedem", 8: "osiem", 9: "dziewięć"}
        self.tab = []
        czyPolski = (len(self.numer))

        if czyPolski == 9:
            print("Prawidłowy numer")
        else:
            print("Nieprawidłowy numer")
 

    @property
    def wojewodztwo(self):
        if self.numer[0] == "4" and self.numer[1] == "2" \
                or self.numer[0] == "4" and self.numer[1] == "3"\
                or self.numer[0] == "4" and self.numer[1] == "4"\
                or self.numer[0] == "4" and self.numer[1] == "6":
            return "Lodzkie"
        if self.numer[0] == "4" and self.numer[1] == "1":
            return "Świętokrzyskie"
        if self.numer[0] == "9" and self.numer[1] == "4" \
                or self.numer[0] == "9" and self.numer[1] == "1":
            return "Zachodnio-pomorskie"
        else:
            return "Nie posiadam informacji o województwie"
    def __str__(self):
        for i in self.numer:
            self.tab.append(self.slownie[int(i)])
        return "Numer telefonu to: {}".format(" ".join(self.tab))
 
#Co chcemy należy odkomentować:

numer_lodz = NumerTelefonu("426877868")
#numer_piotrkow = NumerTelefonu("446666666")
#numer_sieradz = NumerTelefonu("436666666")
#numer_skierniewice = NumerTelefonu("466666666")
#numer_kielce = NumerTelefonu("416666666")
#numer_koszalin = NumerTelefonu("946666666")
#numer_szczecin = NumerTelefonu("916666666")
 
#Wykorzystanie informacji o wojewodztwie:
#numer_lodz.wojewodztwo
 
#Printowanie polskimi słowami:
print(numer_lodz)
```
